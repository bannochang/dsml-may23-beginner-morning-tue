```python
data = movies.merge(
    directors,
    how = "left",
    left_on = "director_id",
    right_on = "id"
)
```

```python
data.drop(["director_id", "id_y"], axis = 1, inplace = True)
```

```python
def encode_gender(data):
    if data == "Male":
        return 0
    else:
        return 1
```

```python
encode_gender = lambda data: 0 if data == "Male" else 1
```

```python
data["gender"] = data["gender"].apply(encode_gender)
```

```python
# Risky Movies
# Budget > Avg revenue of the director
```

```python
def risky_func(x):
    x["risky"] = (x["budget"] - x["revenue"].mean()) >= 0
    return x
```

```python
data_risky = data.groupby("director_name").apply(risky_func)
```
