### Spearman Correlation - 

- Assigns ranks to all data points in cases where the relationship between variables is not linear.
- Converts non - linear relationships to linear relationships
- Converts unranked (nominal) data to ranked (ordinal) data

```python
from scipy.stats import pearsonr, spearmanr
```

```python
pearsonr(df_hw["Height"], df_hw["Weight"])
```

```python
spearmanr(df_hw["Height"], df_hw["Weight"])
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
```

```python
spearmanr(x, y)
```

```python
df = pd.read_csv("concrete_data.csv")
```

```python
df.head()
```

```python
plt.figure(figsize = (10, 8))
correlation_values = df.corr(method = "pearson")
sns.heatmap(
    correlation_values,
    vmax = 0.6,
    linewidths = 0.01,
    square = True,
    annot = True,
    cmap = "RdBu",
    linecolor = "black"
)
plt.title("Correlation between Concrete features")
plt.show()
```
