```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(1, 7, 100000)
# The code above generates 10000 random uniformly distributed values
# in the range 1 and 6 (7 is not included)
```

```python
plt.hist(
    data,
    bins = 6,
    align = "mid",
    edgecolor = "black",
    color = "skyblue",
    rwidth = 0.8
)

plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Histogram for a Die Roll Simulation (Uniform Distribution)")

plt.show()
```
